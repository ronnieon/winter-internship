```python
df2 = pd.read_csv("Data-cleaning-for-beginners-using-pandas.csv")
```

```python
df2.columns = df2.columns.str.lower().str.replace(" ","_")
df2.columns
```

```python
df2.easy_apply = df2.easy_apply.replace("TRUE", 1)
```

```python
df2.head()
```

```python
df2.easy_apply = df2.easy_apply.replace("-1",0)
df2.head()
```

```python
df2.easy_apply = df2.easy_apply.astype('bool')
df2.head()
```

```python
df2
```

```python
df2.isna().value_counts()
```

```python
df2.isnull().value_counts()
```

```python
df2.isnull().sum()
```

```python
avg_age = df2.age.mean()
avg_age
```

```python
df2.age = df2.age.replace("NaN", avg_age)
```

```python
df2.isna().sum()
```

```python
df2
```

```python
df2.age = df2.age.fillna(avg_age)
df2.age = df2.age.astype('int64')
df2.isna().sum()
```

```python
rating_avg = df2.rating.mean()
```

```python
rating_avg
```

```python
df2.rating = df2.rating.fillna(0)
```

```python
df2.isna().sum()
```

```python
df2.salary = df2.salary.replace("k", "000")
```

```python
df2.head()
```

```python
for i in df2.salary:
    j = i.replace("k", "000")
    j = j.replace("$","")
    df2.salary = df2.salary.replace(i,j)
df2.head()
```

```python
df2.head()
```

```python
df2.established = df2.established.replace(-1, 2004)
```

```python
df2.head()
```

```python
df2.rating = df2.rating.replace(-1.0, 0.0)
df2.head()
```

```python
df2
```

```python
for i in df2.location:
    j = i.replace(",In", "")
    j = j.replace(" In", "")
    j = j.replace(" Aus", "")
    j = j.replace(" Ny", "")
    j = j.replace(",Ny", "")
    df2.location = df2.location.replace(i, j)
```
