```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
trains = pd.read_csv("train.csv")
```

```python
trains["Postal Code"].dtype
```

```python
trains["Postal Code"] = trains["Postal Code"].astype(int)
```

```python
trains.isna().sum()
```

```python
trains = trains.fillna(0)
```

```python
trains["Postal Code"] = trains["Postal Code"].astype(int)
```

```python
trains["Category"].value_counts()
```

```python
trains["Sales"] = trains["Sales"].astype(int)
```

```python
trains.describe()
```

```python
trains["Row ID"].dtype
```

```python
sns.histplot(trains["Sales"], kde = True)
plt.xlabel("Sales")
plt.title("Sale Histogram")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.pairplot(trains, hue="Region", size=3)
plt.show()
```

```python
sns.set_style("darkgrid")
sns.FacetGrid(trains, hue = "Category") \
    .map(sns.distplot,"Sales") \
    .add_legend()
plt.show()
```
