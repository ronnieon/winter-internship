```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars = pd.read_csv("cars_ds_final.csv")
```

```python
cars.shape
```

```python
cars.head()
```

```python
cars.isna().sum()
```

```python
cars["Car_name"] = cars['Make'] + '-' + cars['Model']
```

```python
cars.head()
```

```python
cars["Car_name"].value_counts()
```

```python
cars["Ex-Showroom_Price"] = cars["Ex-Showroom_Price"].str.replace("Rs.", "")
cars["Ex-Showroom_Price"] = cars["Ex-Showroom_Price"].str.replace(",", "")
```

```python
cars.head()
```

```python
cars.describe()
```

```python
cars['Ex-Showroom_Price'] = cars['Ex-Showroom_Price'].astype(int)
```

```python
sns.displot(cars["Ex-Showroom_Price"])
```

```python
cars.plot(kind = "scatter", x = "Valves_Per_Cylinder", y = "Seating_Capacity")
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(cars, hue = "Make") \
    .map(plt.scatter,"Valves_Per_Cylinder","Number_of_Airbags") \
    .add_legend()
plt.show()
```

```python
sns.histplot(cars["Ex-Showroom_Price"], kde = True)
plt.xlabel("Ex showroom Price")
plt.title("Price Histogram")
plt.show()
```

```python
cars["Valves_Per_Cylinder"].plot(kind='kde')
```

```python
sns.set_style("whitegrid")
sns.pairplot(cars, hue="Make", size=3)
plt.show()
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(cars, hue = "Make") \
    .map(sns.distplot,"Seating_Capacity") \
    .add_legend()
plt.show()
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(cars, hue = "Make") \
    .map(sns.distplot,"Doors") \
    .add_legend()
plt.show()
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(cars, hue = "Make") \
    .map(sns.distplot,"Cylinders") \
    .add_legend()
plt.show()
```
